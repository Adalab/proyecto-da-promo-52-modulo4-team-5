```python
import pandas as pd
```

```python
def abrir (nombre_archivo):
    '''
    Abre el archivo csv
    '''
    df1 = pd.read_csv(nombre_archivo, on_bad_lines='skip', index_col=0)
    return df1
```

```python
df = abrir('true_cost_fast_fashion.csv')
df.head()
```

```python
def eda_basico(df):

    print('üîÜEjemplo de datos del df:')
    print('________________________________________________________________________________________________________')
    print('‚òòÔ∏èprimeras 3 filas:')
    display(df.head(3))
    print('________________________________________________________________________________________________________')
    print('üåü√∫ltimas 3 filas:')
    display(df.tail(3))
    print('________________________________________________________________________________________________________')
    print('üå∏Muestra aleatoria de 3 filas:')
    display(df.sample(3))
    print('________________________________________________________________________________________________________')

    print('üåªN√∫mero de Filas:')
    display(df.shape[0])
    print('________________________________________________________________________________________________________')

    print('üå±N√∫mero de Columnas:')
    display(df.shape[1])
    print('________________________________________________________________________________________________________')

    print('üåºInformaci√≥n de la tabla:')
    display(df.info())
    print('________________________________________________________________________________________________________')

    print('‚öúÔ∏èNombre de las columnas:')
    display(df.columns)
    print('________________________________________________________________________________________________________')

    print('üî¢Descripci√≥n de los datos num√©ricos:')
    display(df.describe().T)
    print('________________________________________________________________________________________________________')

    print('üî°Descripci√≥n de los datos no-num√©ricos:')
    display(df.describe(include='object').T)
    print('________________________________________________________________________________________________________')

    print('üîµSaber si hay datos √∫nicos:')
    display(df.nunique())
    print('________________________________________________________________________________________________________')

    print('üöÄ Valores √∫nicos en columnas categ√≥ricas:')
    for col in df.select_dtypes(include='object'):
        print(f'üî∏ {col}')
        print('-----------------------------')
        print(df[col].unique())
        print('________________________________________________________________________________________________________')

    print('‚ùåQue datos son nulos por columnas:')
    display(df.isnull().sum())
    print('________________________________________________________________________________________________________')

    print('‚ö†Ô∏èFilas duplicadas:')
    total_duplicados = df.duplicated().sum()
    if total_duplicados > 0:
        print(f'cantidad de duplicados: {total_duplicados}')
        print('Primeros duplicados')
        display(df[df.duplicated()].head(3))
    else:
        print('No hay duplicados')
    print('________________________________________________________________________________________________________')
    return
```

```python
eda_basico(df)
```
