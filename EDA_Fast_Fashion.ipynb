```python
import pandas as pd
```

```python
def abrir (nombre_archivo):
    '''
    Abre el archivo csv
    '''
    df1 = pd.read_csv(nombre_archivo, on_bad_lines='skip', index_col=0)
    return df1
```

```python
df = abrir('true_cost_fast_fashion.csv')
df.head()
```

```python
def eda_basico(df):

    print('🔆Ejemplo de datos del df:')
    print('________________________________________________________________________________________________________')
    print('☘️primeras 3 filas:')
    display(df.head(3))
    print('________________________________________________________________________________________________________')
    print('🌟últimas 3 filas:')
    display(df.tail(3))
    print('________________________________________________________________________________________________________')
    print('🌸Muestra aleatoria de 3 filas:')
    display(df.sample(3))
    print('________________________________________________________________________________________________________')

    print('🌻Número de Filas:')
    display(df.shape[0])
    print('________________________________________________________________________________________________________')

    print('🌱Número de Columnas:')
    display(df.shape[1])
    print('________________________________________________________________________________________________________')

    print('🌼Información de la tabla:')
    display(df.info())
    print('________________________________________________________________________________________________________')

    print('⚜️Nombre de las columnas:')
    display(df.columns)
    print('________________________________________________________________________________________________________')

    print('🔢Descripción de los datos numéricos:')
    display(df.describe().T)
    print('________________________________________________________________________________________________________')

    print('🔡Descripción de los datos no-numéricos:')
    display(df.describe(include='object').T)
    print('________________________________________________________________________________________________________')

    print('🔵Saber si hay datos únicos:')
    display(df.nunique())
    print('________________________________________________________________________________________________________')

    print('🚀 Valores únicos en columnas categóricas:')
    for col in df.select_dtypes(include='object'):
        print(f'🔸 {col}')
        print('-----------------------------')
        print(df[col].unique())
        print('________________________________________________________________________________________________________')

    print('❌Que datos son nulos por columnas:')
    display(df.isnull().sum())
    print('________________________________________________________________________________________________________')

    print('⚠️Filas duplicadas:')
    total_duplicados = df.duplicated().sum()
    if total_duplicados > 0:
        print(f'cantidad de duplicados: {total_duplicados}')
        print('Primeros duplicados')
        display(df[df.duplicated()].head(3))
    else:
        print('No hay duplicados')
    print('________________________________________________________________________________________________________')
    return
```

```python
eda_basico(df)
```
